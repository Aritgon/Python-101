### Assignment 14: Counting with Dictionaries

```python
# exercises revolving around dictionaries and functions.

sales = [
    {'product': 'Laptop', 'price': 1200, 'quantity': 2},
    {'product': 'Mouse', 'price': 25, 'quantity': 5},
    {'product': 'Keyboard', 'price': 75, 'quantity': 3},
    {'product': 'Monitor', 'price': 300, 'quantity': 1}
]

# 1. Calculate total revenue (price * quantity for each item)
def total_rev(dict):
    total_cost = 0

    for items in dict:
        revenue = items['price'] * items['quantity'] # revenue per product.
        total_cost += revenue # total_cost after each iteration.
    return total_cost

print(f"The total price of all products would be : {total_rev(sales)}")

# 2. Find the product with highest revenue
def highest_price(dict):
    """getting the highest price among all products."""

    max_price = 0 # assuming the first value will be the maximum value of the dictionary.
    
    for items in dict:
        if items['price'] > max_price:
            max_price = items['price']
    return max_price

print(highest_price(sales))

# 3. Calculate average price across all products
def avg_price(dict):
    """getting the average price of all prices from the dictionary"""

    price_list = []

    for items in dict:
        price_list.append(items['price'])

    average_price = sum(price_list) / len(price_list)
    return average_price

# print(f"average price of the whole dictionary is : {avg_price(sales)}")
```

```python
# Reverse a list WITHOUT using .reverse() or [::-1]

# this is two pointer method in lists.
# list = [1, 2, 3, 4, 5]
list = [19, 23, 34, 44, 50, 62]

# two pointer method.
def reverse_list(arr):

    left = 0 # assuming value, later in the loop this will work as an index identifier.
    right = len(arr) - 1 # example, if length 5, right value will be 4.
    
    while left < right: 
        # accessing list elements and swapping with each other.
        arr[left], arr[right] = arr[right], arr[left]
        left+=1 # incrementing left position because for the question, that position is placed with the maximum value.
        right-=1 # decreasing right position in the list as it is filled with the smallest element.
    return arr

print(reverse_list(list))
```
