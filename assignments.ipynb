```python
# Write a program that asks the user to input numbers until they input 0. The program should print the sum of all the input numbers.

# for this problem, we need an unbreakable loop.

sum = 0

while True:
    number = int(input("Enter a number: "))
    if number == 0:
        break
    else:
        sum += number
print(f"sum of all given numbers are {sum}")
```

```python
# Write a program that prints all the numbers from 1 to 10 except 5 using a for loop and continue statement.

for i in range(1, 11):
    if i == 5:
        continue
    print(i)
```

```python
# Write a program that defines an empty function using the pass statement.

def empty_func(num):
    for i in range(1, num+1):
        pass # pass is an null function. It doesn't do anything.
        print(i) 

ask = 5
empty_func(ask)
```

```python
# Write a program that asks the user to input a number and prints all the even numbers from 1 to that number using a for loop.

num = int(input("Enter the number: "))

for i in range(1, num+1):
    if i % 2 != 0:
        continue
    print(i)
```

```python
# Write a program that calculates the factorial of a number input by the user using a while loop.

# def factorial(n):
#     if n <= 1:
#         return 1
#     else:
#         result = n * factorial(n-1)
#         return result
# print(factorial(5))

# using while loop.
num = int(input("Enter the number: "))
fact = 1
count = 1

while count <= num:
    fact *= count
    count+=1
print(f"the factorial of {num} is {fact}")
```

```python
# Write a program that calculates the sum of the digits of a number input by the user using a while loop.
number = 123
sum_total = 0

while number > 0: # condition of the loop, until 0 or null appears, it goes on.
    digits = number % 10 # taking the modulus score, which is the value after a division. such as 123 % 10 = 3
    sum_total += digits # adding the left over digit.
    number = number // 10 # iterator, floor division leave no left over value. 123 // 10 will be 12 and we already added the left over value in the prev resulting variable. in the next loop, 12 // 10 will 1, 2 will be added on the sum_total variable after going through the operation of 12 % 10 = 2.
# print(f"{sum_total}")
```

```python
# Write a program that checks if a number input by the user is a prime number using a for loop.

number = 12
is_prime = True

if (number <= 1):
    is_prime = False

for i in range(2, number):
    if number % i == 0:
        is_prime = False
        break

if is_prime:
    print(f"{number} is a prime number.")
else:
    print(f"{number} is not a prime number.")
```

```python
# Write a program that prints the first n Fibonacci numbers, where n is input by the user.
num = 5
a, b= 0,1
count = 0

while count < num:
    print(a, end= " ")
    a, b = b, a+b
    count+=1
```
