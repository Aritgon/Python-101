```python
# Create a list of random numbers and sort it in ascending and descending order. Remove the duplicates from the list and print the modified list.

random_list = [1, 2, 2, 3, 14, 5, 34, 2, 2323, 56.7, 67, 1, 21, 12, 15, 13]

sorted_list = sorted(random_list) # sorted list.
# print(sorted_list)
sorted_in_reverse_list = sorted(random_list, reverse= True) # sorting in reverse order.
# print(sorted_in_reverse_list)

# using set to remove the duplicates from the list. 
# duplicated_free_list = set(random_list) # set prefers unique values, and also it is immutable.
# print(type(duplicated_free_list), duplicated_free_list)
# duplicated_free_list.add("soham")
# print(duplicated_free_list)

# using for loop to delete duplicate values.
duplicated_free_list = []
for i in random_list:
    if i in duplicated_free_list:
        continue
    else:
        duplicated_free_list.append(i)
print(random_list)
print(duplicated_free_list)
```
