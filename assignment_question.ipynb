**Question 5:** Write a Python program to calculate the factorial of a number.

```python
# Your code here
def factorial(n):
    if n == 0:
        return 1
    else:
        result = n * factorial(n-1)
    return result

number = int(input("Enter the number to calculate factorial of the number: "))
print(factorial(number))
```

**Question 17:** Write a Python program to check if a year is a leap year.

```python
# Your code here
year = int(input("Enter a year to check: "))

if (year % 400 == 0) or (year % 4 == 0 and year % 100 != 0):
    print(f"{year} is a leap year")
else:
    print(f"{year} is not a leap year")
```

**Question 18:** Write a Python program to reverse a string.

```python
# Your code here
def reverse_str(str):
    temp = ""

    for char in str:
        temp = char + temp
    return temp

name = "arit"
# print(reverse_str(name))

def reverse_str_2nd(string):
    result = string[::-1]
    return result

print(reverse_str_2nd(name))
```

**Question 19:** Write a Python program to check if a string is a palindrome.

```python
# Your code here

def palindrome_checker(str):
    temp = ""

    for char in str:
        temp = char + temp
    
    if temp == str:
        print(f"{str} is a palindrome word.")
    else:
        print(f"{str} is not a palindrome word.")


word = "madam"
palindrome_checker(word)
```
