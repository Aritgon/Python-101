```python
## else
## The else statement executes a block of code if the condition in the if statement is False.

age=16

if age>=18: # if block always evaluates true as condition.
    print("You are eligible for voting")
else: # this block will run if the condition is in the if block is false.
    print("You are a minor")
```

```python
## elif
## The elif statement allows you to check multiple conditions. It stands for "else if"

age=17

if age<13:
    print("You are a child")
elif age<18:
    print("You are a teenager")
else:
    print("You are an adult")
```

```python
## Nested Condiitonal Statements

# You can place one or more if, elif, or else statements inside another if, elif, or else statement to create nested conditional statements.

## number even ,odd,negative

num = int(input("Enter a number: "))

if num > 0:
    if num % 2 == 0:
        print(f"{num} is an even number.")
    else:
        print(f"{num} is an odd number.")
else:
    print(f"{num} is a negative number.")
```

```python
## Practical Examples

## Determine if a year is a leap year using nested condition statement

year = int(input("Enter a year: "))

if year % 4 ==0:
    if year % 100 == 0:
        if year % 400 == 0:
            print(f"{year} is a leap year.")
        else:
            print(f"{year} is not a leap year.")
    else:
        print(f"{year} is a leap year.") # because anything divisible by 4 but not divisible by 100 is considered as leap year.
else:
    print(f"{year} is not a leap year")
```

```python
## Assignment
## Simple Calculator program
# Take user input
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))
operation = input("Enter operation (+, -, *, /): ")

# Perform the requested operation
if operation == "+":
    print(num1 + num2)
elif operation == "-":
    print(num1 - num2)
elif operation == "/":
    if num2 != 0:
        print(num1 / num2)
    else:
        print("Zero division error.")
elif operation == "*":
    print(num1 * num2)
else:
    print("no operator given!")
```

```python
### Determine the ticket price based on age and whether the person is a student.
# Ticket pricing based on age and student status

# conditions
# age < 5, price free.
# age <= 12, price $10
# age <= 17 and if student, then price $12, of not student, price $15
# age <= 64 and if student, then price $18, if not student, price $25
# if no age mentioned, than go on with $20 rupees by default.


# Take user input
age = int(input("Enter the age: "))
is_student = input("Are you a student: (yes/no): ").lower()

if age<5:
    print("price is free.")
elif age <= 12:
    print("price is $12")
elif age <= 17:
    if is_student == 'yes':
        print("price is $12")
    else:
        print("price is $15")
elif age < 64:
    if is_student == 'yes':
        print("price is $18")
    else:
        print("price is $25")
else:
    print("price is $20")
```

#### Complex Example 3: Employee Bonus Calculation

Calculate an employee's bonus based on their performance rating and years of service.

```python
# Employee bonus calculation

# conditions 
# experience is greater or equals to 10 and performance is greater than 4, then bonus is 20%, otherwise bonus is 15%
# experience is greater or equals to 5 and performance is greater than 4, then bonus is 15%, if performance is below 4, then bonus of 10%, other wise 5% bonus.
# if experience is below 5 and performance is greater or equals to 4.5, then bonus of 12%, other wise zero bonus.

exp_year = int(input("Enter the exp. year: "))
performance_rating = float(input("Enter your performance rating (1 to 5): "))

bonus = 0

if exp_year >= 10:
    if performance_rating >= 4:
        bonus = 20
    else:
        bonus = 15
elif exp_year >= 5:
    if performance_rating >= 4:
        bonus = 15
    elif performance_rating <= 4 and performance_rating >= 2:
        bonus = 10
    else:
        bonus = 5
elif exp_year < 5 and performance_rating >= 4.5:
    bonus = 12
else:
    bonus = 0

salary = float(input('Enter your current salary: '))
result = salary * bonus / 100

print(f"your bonus will be {result}")
```

## Complex Example 4: User Login System
A simple user login system that checks the username and password.

```python
# User login system

# Predefined username and password
stored_username = "admin"
stored_password = "password123"

# Take user input
username = input("Enter your username: ")
password = input("Enter your password: ")

if username == stored_username:
    if len(stored_password) != len(password):
        print(f"You typed a {len(password)} length password, where your password as {len(stored_password)} length")
    elif stored_password == password:
        print("Your credentials matched.")
    else:
        print("wrong password but the length of the password is correct. Think more!")
else:
    print("wrong username!")
```
