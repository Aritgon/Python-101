```python
# Category wise total sales count.
category_total = (
    df.groupby('Category', as_index=False).agg
    (
        total_sales = ('Sales','sum')
    )
)

category_total.plot(x = 'Category', y = 'total_sales', figsize=(6, 6), kind = 'bar', title = "total amount by category")
```

```python
# product wise total sales count.
df.groupby('Product')['Sales'].sum().reset_index()
```

```python
df.columns
```

```python
# changing column names.
df.rename(columns={'Date' : 'date', 'Category' : 'category', 'Value' : 'value', 'Product': 'product', 'Sales': 'sales', 'Region': 'region'}, inplace=True)
```

```python
df.columns
```

```python
df.head()
```

```python
# extracting year, month, weekname, week_period, day_name from date column
# df['date'].dtype #O -> object, need to format it to proper data-type like data.
# df['date'] = pd.to_datetime(df['date'], format = 'mixed')
df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month
df['day_name'] = df['date'].dt.day_name
df['week_period'] = df['date'].dt.to_period('W')
```

```python
df.head()
```

```python
df['day_name'].dtype
```

```python
# dropping uncertain column
df.drop(columns= ['day_name'], inplace=True)
```

```python
df
```

```python
# week_period wise avg sales distribution.
weekly_avg_sales_dist = (
    df.groupby('week_period', as_index=False).agg
    (
        avg_sales = ('sales', 'mean')
    )
)

weekly_avg_sales_dist.sort_values(by = 'week_period', ascending=False).plot(x = 'week_period', y = 'avg_sales', kind = 'barh', figsize=(10, 6), title="weekly sales distribution")
```

```python
# Region wise avg sales.
df.groupby('region')['sales'].mean().reset_index().sort_values(by = 'sales', ascending=False).plot(x = 'region', y= 'sales', kind = 'bar', figsize=(10, 6) ,title="avg sales per region")
```

```python
# total sales volume per month.
df.groupby('month')['sales'].count().reset_index()
```

```python
# statistical summary 
df.describe()
```

```python
# adding week column
df['weak'] = df['date'].dt.isocalendar().week
```

```python
df.head()
```

```python
# weekly avg sales distribution.

df.groupby('week')['sales'].mean().reset_index().sort_values(by = 'week').plot(x = 'week', y = 'sales', kind = 'pie', figsize=(10,7), title="avg weekly sales distribution")
```

```python
# changing the week name.
df.rename(columns = {'weak':'week'},inplace=True)
```

```python
# checking which product has the most selling count?
df.groupby('product')['sales'].count().reset_index().sort_values(by = 'sales')
```

```python
df
```

```python
# each product average selling value.
df.groupby('product')['sales'].mean().reset_index().sort_values(by = 'sales', ascending=False)
```

```python
# day wise selling trend.
df.groupby('date')['sales'].mean().reset_index().sort_values(by = 'date').plot(x = 'date', y='sales', kind = 'line', figsize=(20, 7), title="day wise selling trend")
```

```python
# region wise category performance.
# avg sales value

grouped_sales = (
    df.groupby(['region', 'category'], as_index=False).agg(
        avg_sales = ('sales', 'mean'),
        total_count = ('sales', 'count')
    )
)

grouped_sales['regional_category'] = (grouped_sales['region'] + " - " + grouped_sales['category'])

grouped_sales.reset_index().plot(x = 'regional_category', y=['avg_sales', 'total_count'], kind = 'bar', figsize=(10, 8), title='region wise category performance')
```
