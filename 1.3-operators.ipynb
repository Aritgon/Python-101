```python
#greater than or equal to
# as one of the operator is evaluating the condition.
n1 = 45
n2 = 45
print(n1>n2) # false
print(n1 >= n2) # as True
print(n1 < n2) # false
print(n1 <= n2) # true
```

```python
#less than or equal to
number1=44
number2=45

print(number1<=number2) # true.
print(number2 >= number1) # true.
```

#### Logical Operators

- AND -> It strictly checks for both the condition to be True or False.
- OR -> It checks if any of the condition is right or True. 
- NOT -> It converts condition results (True to False and vice versa). Specially, it changes the output of any condition to opposite.

```python
## And ,Not,OR

x = True
y = True

# print(x==y)
# print(x!=y)

result = x and y
# print(result)
result1 = x or y
# print(result1)

result2 = not(x and y) # it converts the actual datatype or bool value to opposite (True -> False and vice versa)
print(result2)
```

```python
x = False
y = True

result = x and y
print(result)
```
