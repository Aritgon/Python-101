#### Sets
Sets are a built-in data type in Python used to store collections of unique items. They are unordered, meaning that the elements do not follow a specific order, and they do not allow duplicate elements. Sets are useful for membership tests, eliminating duplicate entries, and performing mathematical set operations like union, intersection, difference, and symmetric difference.

```python
## MAthematical Operation
set1={1,2,3,4,5,6}
set2={4,5,6,7,8,9}

union_set = set1.union(set2) # returns all elements, unionize with known element of both sets.
print(union_set)

intersect_set = set1.intersection(set2) # returns only the common values.
print(intersect_set)

# intersected_update_set = set1.intersection_update(set2)
# print(intersected_update_set)
# set1.intersection_update(set2)
# print(set1)

# difference -> outputs the different elements that are present in the first set, while completely ignores diff elements in the set if present.
# difference_set = set1.difference(set2)
difference_set = set2.difference(set1)
print(difference_set)

# symmetric difference -> returns unique and different elements from both sets.
symm_diff_set = set1.symmetric_difference(set2)
print(symm_diff_set)


# all methods recap - 
# union -> returns all elements along with common elements.
# intersect -> returns only the common values among set.
# difference -> returns unique and different elements from the first set while ignores all element of second set.
# symmetric_difference -> returns unique and different elements from all sets.
```
